# **PrediCare - AI Breast Cancer Diagnosis**

**Breast cancer**, the most prevalent cancer in women globally, is characterized by uncontrolled cell growth in breast tissue, underscoring the paramount importance of early detection for successful treatment and improved outcomes.

The diagnostic protocol for breast cancer typically involves a series of routine check-ups, including physical examinations, imaging tests, and further diagnostics such as biopsies, gene sequencing, or tumor board evaluations.

Two primary imaging techniques utilized for early detection are mammography and sonography. Mammography, employing X-ray radiation, is the standard for capturing detailed breast tissue images and is widely used for routine screening. However, its effectiveness can be limited in dense breast tissue, and its application is restricted for X-ray-sensitive populations, such as pregnant women. Additionally, accessibility to mammography devices varies across countries.

**Sonography**, or ultrasound, utilizes sound waves to generate images of breast tissue. It excels in distinguishing between solid masses and fluid-filled cysts, presenting a radiation-free alternative suitable for repeated examinations and pregnant women.

Despite its advantages, sonography presents challenges, relying heavily on operator skill and experience and occasionally resulting in false diagnoses. Various characteristics in the ultrasound image, such as tumor shape, margins, echogenicity, and shadowing, must be considered for accurate analysis, emphasizing the need for skilled professionals in the diagnostic process.

To enhance the accuracy of sonography, ongoing research explores the application of **convolutional neural networks**. These advanced algorithms have the potential to support image analysis, improving diagnostic reliability and minimizing the risk of overlooking potential malignancies.

In our project, we employ image recognition procedures on a breast ultrasound dataset, aiming to provide promising classifications for normal, benign, and malignant cases. The dataset, collected in 2018 from women aged 25 to 75, comprises 780 PNG-format images with an average size of 500x500 pixels. Ground truth images accompany the original images, categorized into three classes: normal, benign, and malignant, totaling 600 female patients.

### **Baseline Model Creation**

As a baseline model, we decided to use a classical machine learning algorithm. The **Support Vector Machine (SVM)** is a powerful algorithm that excels in image classification tasks. SVM operates by identifying a hyperplane that best separates different classes in the feature space, aiming to maximize the margin between these classes. It can handle high-dimensional data, such as images represented as pixel values

The algorithm's simplicity and interpretability make it well-suited for initial modeling in the image classification tasks.
Its performance as a baseline model provides a benchmark for further algorithms.


```python
# Extracting the images paths and labels
images, labels = read_data(data_dir_raw)

print(len(images))
print(len(labels))
```

After extracting the images and the labels (normal, benign, malignant), a pandas data frame is created with the image paths and their corresponding labels.

```python
# Creating a pandas data frame
df = pd.DataFrame({'images': images, 'labels': labels})

# Perform train-test-split
train_data, test_data = train_test_split(df, test_size = 0.2, random_state=42, stratify=df['labels'])
train_data = train_data.reset_index(drop=True)
test_data = test_data.reset_index(drop=True)
```

The data necessitates pre-processing to make it compatible with the SVM algorithm. Standardizing the size of all images is crucial, and the transformation from a 3-dimensional array to a 1-dimensional array is essential. This pre-processing step is applied to both the training and test sets, ensuring uniformity and facilitating effective input for the SVM algorithm.

```python
# Data processing for train data set
height, width, channels = 32, 32, 3

# Initialize empty arrays for X_train and y_train
X_train = np.zeros((len(train_data), height * width * channels), dtype=np.uint8)
y_train = np.zeros(len(train_data), dtype=np.uint8)


# Iterate through the DataFrame to populate X_train and y_train
for idx, row in train_data.iterrows():
    image_path = row['images']
    label_train = row['labels']

    # Load and resize the image
    image = cv2.imread(image_path)
    image = cv2.resize(image, (width, height))
    image_flattened = image.flatten()  # Make a 1-dimensional array

    X_train[idx] = image_flattened
    y_train[idx] = label_train

print(X_train.shape)
print(len(y_train))
```

```python
# Data processing for test data set
height, width, channels = 32, 32, 3

# Initialize empty arrays for X_test and y_test
X_test = np.zeros((len(test_data), height * width * channels), dtype=np.uint8)
y_test = np.zeros(len(test_data), dtype=np.uint8)

# Iterate through the DataFrame to populate X_test and y_test
for idx, row in test_data.iterrows():
    image_path = row['images']
    label_test = row['labels']

    # Load and resize the image
    image = cv2.imread(image_path)
    image = cv2.resize(image, (width, height))
    image_flattened = image.flatten()  # Make a 1-dimensional array

    X_test[idx] = image_flattened
    y_test[idx] = label_test

print(X_test.shape)
print(len(y_test))
```

Having completed the pre-processing of the images, we are now ready to input them into the algorithm.

```python
# Train the SVM model
svm = SVC()
svm.fit(X_train, y_train)
```

The trained SVM baseline model is now employed for predictions on the test dataset. To assess the performance of the model, both a classification report and a confusion matrix have been generated, providing comprehensive insights into the model's predictive accuracy and potential areas for improvement.





```python
# Make predictions on the test set
y_pred = svm.predict(X_test)
```

```python
#Calculate the accuracy and display the classification report
categories = {'normal': 0, 'benign': 1, 'malignant': 2}
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy SVM on Test set:", accuracy)
print("Classification Report on Test set:")
print(classification_report(y_test, y_pred, target_names=categories))
```

```python
# Create a confusion matrix and visualize it
confusion = confusion_matrix(y_test, y_pred)
categories = {'normal': 0, 'benign': 1, 'malignant': 2}

plt.figure(figsize=(8, 6))
sns.heatmap(confusion, annot=True, fmt='d', cmap='Oranges', xticklabels=categories, yticklabels=categories)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix: SVM baseline model with original images')
plt.show()
```
